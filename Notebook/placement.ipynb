```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('placement.csv')
df
```

# Companies Visited with highest package

```python
a = 'CSE'
```

```python
a_df = df[df['BRANCH'] == a]
a_df
```

```python
x = a_df['COMPANY_NAME']
y = a_df['HIGHEST_PACKAGE']

plt.bar(x,y,color='blue')
plt.xlabel('Companies Visited for {}'.format(a))
plt.ylabel('Highest Package')
plt.rcParams["figure.figsize"] = [10,9]
plt.show()
```

# Companies visited with total appeared vs total placed

```python
a = 'CSE'
```

```python
a_df = df[df['BRANCH']=='CSE']
a_df
```

```python
tot_app = a_df['TOTAL_APPEARED']
tot_plac = a_df['TOTAL_PLACED']
x = a_df['COMPANY_NAME']


width = 0.35

plt.bar(x,tot_app,width,label='Total Appeared')
plt.bar(x,tot_plac,width,label='Total Placed')
plt.legend(loc='best')
plt.show()
```
